Divide o arquivo csv das uninas nucleares em um arquivo por pais, atuando com os dados dentro do google drive.

```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('/content/drive/MyDrive/NUCLEAR/usinas_nucleares.csv')
dados
```

```python
paises = dados['country_long'].unique()
paises = paises.tolist()
paises
```

```python
colunas = ['country_long', 'commissioning_year']
for pais in paises:
  dados_pais = dados[dados['country_long'] == pais].copy()
  dados_pais.drop(columns = colunas, inplace=True)
  dados_pais.to_csv('/content/drive/MyDrive/NUCLEAR/nuclear_'+pais+'.csv', index=False)
```
