Faz leituras de diversos arquivo html obtidos no site da ONU (https://pris.iaea.org/PRIS/CountryStatistics/CountryStatisticsLandingPage.aspx) com infomações sobre reatres nucleares, realiza os tratamentos necessario e ao fim gera um dicionario com as infrmações compiladas por pais, que poteriomente foi utilizado como fonte de dados para alimentar a elabboração de um site.

Os arquivos html pode sem encontrados nesse endereço: https://drive.google.com/drive/folders/1C6zwn1ys8mbi1e7MCiJjAOzhv9EuMPED?usp=sharing



```python
import pandas as pd
import os
import re
```

```python
url_codigos_paises = '/content/drive/MyDrive/ALURA/codigo_paises.csv'
```

```python
codigo_paises = pd.read_csv(url_codigos_paises)
codigo_paises
```

```python
url_usinas = '/content/drive/MyDrive/ALURA/paises'
folder = os.listdir(url_usinas)
```

```python
completo = pd.DataFrame(columns=["Name", "Type", "Status", "Location", "Reference Unit Power  [MW]", "Gross Electrical Capacity [MW]", "First Grid Connection", "Pais", "Site"])
completo
```

```python
tag = []
for file in folder:
  path_usina = os.path.join(url_usinas, file)
  dados = pd.read_html(path_usina)
  dados_usinas = dados[1]
  pais_cod = path_usina[-8:-5]
  onu_cod = path_usina[-11:-9]

  for row in codigo_paises.itertuples():
    if row.Trigrama == pais_cod:
      pais = row.Entidade

  for item in [pais, onu_cod, pais_cod]:
    tag.append(item)

  dados_usinas['Pais'] = pais
  dados_usinas['Site'] = 'https://pris.iaea.org/PRIS/CountryStatistics/CountryDetails.aspx?current=' + onu_cod

  completo = pd.concat([completo, dados_usinas], ignore_index=True)
```

```python
situacao_pt = ['Operacional', 'Em Contrução', 'Desativada Permanentemente', 'Descomissionamento Concluído', 'Operação Supensa']
```

```python
substituicao = {'Operational': 'Operacional', 'Under Construction': 'Em Contrução', 'Permanent Shutdown': 'Desativado Permanentemente', 'Decommissioning Completed': 'Descomissionamento Concluído', 'Suspended Operation': 'Operação Supensa'}
```

```python
completo['Status'] = completo['Status'].map(substituicao)
completo
```

```python
completo['Status'].unique()
```

```python
paises = completo['Pais'].unique().tolist()
```

```python
df_pais = completo[completo['Pais']=='Brasil']
contagem = df_pais['Status'].value_counts()
contagem['Operacional']
```

```python
list_dict_paises = []

for pais in paises:

  dic_pais = {}

  df_pais = completo[completo['Pais']==pais]

  desc_pais = f'{pais} é uma nação que possui '

  n_reatores = len(df_pais['Name'])
  if n_reatores == 1:
    desc_n_reatores = f'{n_reatores} reator nuclear, localizado em '
  else:
    desc_n_reatores = f'{n_reatores} reatores nucleares, localizados em '

  n_cidades = len(df_pais['Location'].unique())
  if n_cidades == 1:
    desc_cidades = f'{n_cidades} cidade. '
  else:
    desc_cidades = f'{n_cidades} cidades. '

  tipo_reatores = df_pais['Type'].unique().tolist()
  if len(tipo_reatores) == 1:
    if n_reatores ==1:
      desc_tipo = f'O reator é do tipo {tipo_reatores[0]}, sendo que, '
    else:
      desc_tipo = f'Os reatores são do tipo {tipo_reatores[0]}, sendo que, '
  else:
    texto = ' e '.join(tipo_reatores)
    texto = texto.replace('X', 'Desconhecido')
    desc_tipo = f'Os reatores são dos tipos {texto}, sendo que, '

  status_reatores = df_pais['Status'].unique().tolist()
  contagem_status = df_pais['Status'].value_counts()
  list_status = []
  for status in status_reatores:
    n_status = contagem_status[status]
    if status == 'Operacional':
      texto = f'{n_status} em situação {status}'
    elif status == 'Em Contrução':
      texto = f'{n_status} {status}'
    elif status == 'Desativado Permanentemente':
      if n_status == 1:
        texto = f'{n_status} {status}'
      else:
        texto = f'{n_status} Desativados Permanentemente'
    elif status == 'Descomissionamento Concluído':
      texto = f'{n_status} com o {status}'
    elif status == 'Operação Supensa':
      texto = f'{n_status} com a {status}'
    list_status.append(texto)
  desc_status = ' e '.join(list_status) + '.'

  descricao = desc_pais + desc_n_reatores + desc_cidades + desc_tipo + desc_status
  fonte = df_pais['Site'].unique().tolist()

  for i in range(len(tag)):
    if tag[i] == pais:
      tags = tag[i:i+3]

  dic_pais['pais'] = pais
  dic_pais['descicao'] = descricao
  dic_pais['fonte'] = fonte[0]
  dic_pais['tags'] = ', '.join(tags)

  list_dict_paises.append(dic_pais)

  print(desc_tipo)
```
