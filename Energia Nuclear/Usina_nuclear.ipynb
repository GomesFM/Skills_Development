```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/power_plant_database_global.csv', sep=(','));
dados
```

```python
dados_nuclear = dados.query("primary_fuel == 'Nuclear'").copy()
```

```python
dados_nuclear
```

```python
colunas = dados_nuclear.columns
colunas
```

```python
colunas = ['country','primary_fuel', 'other_fuel1', 'other_fuel2', 'url','geolocation_source',
       'other_fuel3','generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015', 'year_of_capacity_data',
       'generation_gwh_2016', 'generation_gwh_2017', 'generation_gwh_2018',
       'generation_gwh_2019', 'generation_data_source',
       'estimated_generation_gwh_2013', 'estimated_generation_gwh_2014',
       'estimated_generation_gwh_2015', 'estimated_generation_gwh_2016',
       'estimated_generation_gwh_2017', 'estimated_generation_note_2013',
       'estimated_generation_note_2014', 'estimated_generation_note_2015',
       'estimated_generation_note_2016', 'estimated_generation_note_2017']
```

```python
dados_nuclear.drop(columns=colunas, inplace=True)
```

```python
dados_nuclear.isna().sum()
```

```python
dados_nuclear['owner'].fillna(dados_nuclear['source'], inplace=True)
```

```python
dados_nuclear.isna().sum()
```

```python
dados_nuclear.duplicated().sum()
```

```python
dados_nuclear.drop(columns='source', inplace=True)
```

```python
dados_nuclear.to_csv('usinas_nucleares.csv', index=False)
```
